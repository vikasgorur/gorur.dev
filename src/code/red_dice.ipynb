```python
import matplotlib.pyplot as plt
from collections import defaultdict

content = open("data/dice-rolls.txt", "r").readlines()
rolls = [int(line.strip()) for line in content if not line.startswith("#")]

counts = defaultdict(lambda: 0)
for roll in rolls:
    counts[roll] += 1

uniform_counts = [24, 24, 24, 24, 24, 24]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the actual counts
ax.bar(counts.keys(), counts.values(), color='blue', label='Actual')

# Overlay the uniform counts
ax.bar(range(1, 7), uniform_counts, color='lightblue', alpha=0.4, label='Uniform')

# Customize the plot
ax.set_xlabel('Dice Roll')
ax.set_ylabel('Count')
ax.set_title('Actual vs Uniform Dice Roll Distribution')
ax.legend()

# Display the plot
plt.show()
```

```python
import numpy as np

# Normalize counts to probabilities
P = np.array(list(counts.values())) / len(rolls)

# Uniform distribution
Q = np.ones_like(P) / len(P)
```

```python

def kl_divergence(p, q):
    return sum(p[i] * np.log(p[i] / q[i]) for i in range(len(p)))
```
